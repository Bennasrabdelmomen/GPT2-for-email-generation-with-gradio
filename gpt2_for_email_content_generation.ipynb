```python
import gradio as gr
import tensorflow as tf
```

```python
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer=GPT2Tokenizer.from_pretrained('gpt2')
model=TFGPT2LMHeadModel.from_pretrained('gpt2', pad_token_id=tokenizer.eos_token_id)
```

```python
def generate_email(prompt, additional_info):
    input_text = f"Subject: ML Night Reminder\n\n{prompt} {additional_info}"
    input_ids = tokenizer.encode(input_text, return_tensors='tf')
    beam_output = model.generate(
        input_ids,
        max_length=200,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True,
        temperature=0.8,
    )
    generated_email = tokenizer.decode(
        beam_output[0], skip_special_tokens=True, clean_up_tokenization_spaces=True
    )
    return generated_email

prompt = (
    "All ML Night participants have to be at the location provided by exactly 7 pm. "
    "All unjustified absences will not be tolerated. Now, provide additional information:"
)

additional_info = "The location is Room 123, and please bring a form of identification."

generated_email = generate_email(prompt, additional_info)
print(generated_email)
```

```python
iface = gr.Interface(
    fn=generate_email,
    inputs=[
        gr.Textbox(placeholder="Enter your prompt here", label="Prompt"),
        gr.Textbox(placeholder="Additional info", label="Additional Info"),
    ],
    outputs=gr.Textbox(placeholder="Model output will appear here", label="Model Output"),
)
```

```python
iface.launch()
```

```python
import gradio as gr
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load the pre-trained model and tokenizer
model_name = "gpt2"  # You can use other models as well
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

def generate_email(prompt, additional_info, subject="Subject"):
    input_text = f"Subject: {subject}\n\n{prompt} {additional_info}"
    input_ids = tokenizer.encode(input_text, return_tensors='pt')
    beam_output = model.generate(
        input_ids,
        max_length=200,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True,
        temperature=0.8,
    )
    generated_email = tokenizer.decode(
        beam_output[0], skip_special_tokens=True, clean_up_tokenization_spaces=True
    )
    return generated_email

# Create a Gradio interface
iface = gr.Interface(
    fn=generate_email,
    inputs=[
        gr.Textbox(placeholder="Enter your prompt here", label="Prompt"),
        gr.Textbox(placeholder="Additional info", label="Additional Info"),
        gr.Textbox(placeholder="Subject", label="Subject"),
    ],
    outputs=gr.Textbox(placeholder="Model output will appear here", label="Model Output"),
)

# Launch the Gradio interface
iface.launch()
```
